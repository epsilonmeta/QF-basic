```python
#Simulating the stock price using geometric brownian motion#
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#drift coefficient
mu = 0.1
#number of steps
n = 100
# time in years
T = 1
#number of sims
M = 500
#initial stock price
S0 = 100
#volatility
sigma = 0.1
```

```python
#simulating GBM Paths
dt = T/ n
#simulating using numpy arryas
##St= np.exp(mu-sigma**2/2)*dt +sigma* np.random.normal(0,np.sqrt(dt),size=(M,n)).T
St = np.exp((mu - sigma ** 2 / 2) * dt + sigma * np.random.normal(0, np.sqrt(dt), size=(M,n))).T  
#include arrays of 1s
St = np.vstack([np.ones(M),St])
#multiply through by S0 and return cumulative product of elements along a given simulation path (Axis = 0)
St = S0 *St.cumprod(axis=0) 
St[1]
```

```python
#Consider time interval in years#
time =np.linspace(0,T,n+1)
# require nump array same shape
tt = np.full(shape=(M,n+1),fill_value =time).T
```

```python
plt.plot(tt,St)
plt.xlabel("Years $t$")
plt.ylabel("Stock Price $(S_t)$")
plt.show()
```
