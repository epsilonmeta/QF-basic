```python
# Calculate returns, expected returns, and covariance matrix
def calculate_portfolio_metrics(prices):
    # Calculate daily returns
    returns = prices.pct_change().dropna()
    
    # Calculate expected returns (annualized)
    mu = returns.mean() * 252
    
    # Calculate covariance matrix (annualized)
    cov_matrix = returns.cov() * 252
    
    return returns, mu, cov_matrix

```

```python
# Portfolio optimization functions
def portfolio_return(weights, expected_returns):
    return np.sum(weights * expected_returns)

def portfolio_volatility(weights, cov_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

def negative_sharpe_ratio(weights, expected_returns, cov_matrix, risk_free_rate=0.05):
    portfolio_ret = portfolio_return(weights, expected_returns)
    portfolio_vol = portfolio_volatility(weights, cov_matrix)
    return -(portfolio_ret - risk_free_rate) / portfolio_vol

```

```python
# Generate efficient frontier
def generate_efficient_frontier(expected_returns, cov_matrix, risk_free_rate=0.05, points=100):
    efficient_portfolios = []
    
    # Get minimum variance portfolio
    min_vol_result = minimize(
        portfolio_volatility,
        np.array([1.0 / len(expected_returns)] * len(expected_returns)),
        args=(cov_matrix,),
        method='SLSQP',
        bounds=tuple((0, 1) for _ in range(len(expected_returns))),
    
    )
    min_vol_return = portfolio_return(min_vol_result['x'], expected_returns)
    min_vol_volatility = portfolio_volatility(min_vol_result['x'], cov_matrix)
    
    # Get maximum return portfolio
    max_return_idx = np.argmax(expected_returns)
    max_return = expected_returns[max_return_idx]
    
    # Generate efficient frontier points
    target_returns = np.linspace(min_vol_return, max_return, points)
    
    for target_return in target_returns:
        result = optimize_portfolio(expected_returns, cov_matrix, risk_free_rate, target_return)
        if result['success']:
            weights = result['x']
            portfolio_ret = portfolio_return(weights, expected_returns)
            portfolio_vol = portfolio_volatility(weights, cov_matrix)
            sharpe = (portfolio_ret - risk_free_rate) / portfolio_vol
            efficient_portfolios.append({
                'Return': portfolio_ret,
                'Volatility': portfolio_vol,
                'Sharpe Ratio': sharpe,
                'Weights': weights
            })
    
    return efficient_portfolios

# Plot efficient frontier
def plot_efficient_frontier(efficient_portfolios, optimized_portfolio, stock_returns, risk_free_rate=0.05):
    plt.figure(figsize=(12, 8))
    
    # Plot efficient frontier
    returns = [p['Return'] for p in efficient_portfolios]
    volatilities = [p['Volatility'] for p in efficient_portfolios]
    plt.plot(volatilities, returns, 'b-', linewidth=3)
    
    # Plot optimized portfolio
    plt.scatter(
        optimized_portfolio['Volatility'],
        optimized_portfolio['Return'],
        marker='*',
        color='r',
        s=300,
        label='Optimal Portfolio'
    )
    
    # Plot individual assets
    for i, stock in enumerate(stock_returns.columns):
        plt.scatter(
            np.sqrt(stock_returns.cov().iloc[i, i] * 252),
            stock_returns.mean()[i] * 252,
            marker='o',
            s=50,
            label=stock
        )
    
    # Plot capital market line
    max_sharpe = max(efficient_portfolios, key=lambda x: x['Sharpe Ratio'])
    x_values = [0, max_sharpe['Volatility'] * 1.5]
    y_values = [risk_free_rate, risk_free_rate + max_sharpe['Sharpe Ratio'] * x_values[1]]
    plt.plot(x_values, y_values, 'g--', label='Capital Market Line')
    
    plt.title('Efficient Frontier with Individual Assets', fontsize=16)
    plt.xlabel('Volatility (Standard Deviation)', fontsize=14)
    plt.ylabel('Expected Return', fontsize=14)
    plt.grid(True)
    plt.legend(loc='best')
    
    return plt

# Main execution function
def main():
    # Set parameters
    lookback_years = 3
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365 * lookback_years)
    risk_free_rate = 0.05  # Adjust based on current risk-free rate
    
    # Get stock data
    prices = stock_data
    
    # Calculate returns and metrics
    returns, expected_returns, cov_matrix = calculate_portfolio_metrics(prices)
    
    # Optimize portfolio (maximize Sharpe ratio)
    result = optimize_portfolio(expected_returns, cov_matrix, risk_free_rate)
    
    if result['success']:
        optimal_weights = result['x']
        
        # Get optimized portfolio metrics
        optimal_return = portfolio_return(optimal_weights, expected_returns)
        optimal_volatility = portfolio_volatility(optimal_weights, cov_matrix)
        optimal_sharpe = (optimal_return - risk_free_rate) / optimal_volatility
        
        optimized_portfolio = {
            'Return': optimal_return,
            'Volatility': optimal_volatility,
            'Sharpe Ratio': optimal_sharpe,
            'Weights': optimal_weights
        }
        
        # Generate efficient frontier
        efficient_portfolios = generate_efficient_frontier(expected_returns, cov_matrix, risk_free_rate)
        
        # Display results
        print("\n==== Optimized Portfolio ====")
        print(f"Expected Annual Return: {optimal_return:.4f} ({optimal_return*100:.2f}%)")
        print(f"Expected Annual Volatility: {optimal_volatility:.4f} ({optimal_volatility*100:.2f}%)")
        print(f"Sharpe Ratio: {optimal_sharpe:.4f}")
        
        # Create a DataFrame for weights
        weights_df = pd.DataFrame({
            'Stock': returns.columns,
            'Weight': optimal_weights
        })
        weights_df = weights_df.sort_values(by='Weight', ascending=False)
        
        # Display top 10 holdings
        print("\n==== Top 10 Holdings ====")
        for i, row in weights_df.head(10).iterrows():
            print(f"{row['Stock']}: {row['Weight']*100:.2f}%")
        
        # Plot efficient frontier
        plt = plot_efficient_frontier(efficient_portfolios, optimized_portfolio, returns, risk_free_rate)
        plt.show()
        
        # Save weights to CSV
        weights_df.to_csv('optimal_portfolio_weights.csv', index=False)
        print("\nOptimal portfolio weights saved to 'optimal_portfolio_weights.csv'")
    else:
        print("Optimization failed:", result['message'])

if __name__ == "__main__":
    main()
```
