```python
current_date = (pd.to_datetime('today') - pd.Timedelta(days=1500)).strftime('%Y-%m-%d')  # 1460 days = 4 years
sd1 = sd[sd.index > current_date] # filter data for last 4 years
# drop columns with all NaN values
sd1.dropna(axis=1, how='any', inplace=True)
sd1_columns =sd1.columns # get column names
sd1
```

```python
sd1.to_csv("prices.csv")
```

```python
def momentum_ratio(df): # function to calculate momentum ratio
    for i in df.columns:
        returns252 = df[i].pct_change(periods=252) # percent change over 252 trading days
        stds252 = returns252.rolling(window=252).std() # standard deviation over 252 trading days
        returns126 = df[i].pct_change(periods=126) # percent change over 126 trading days
        stds126 = returns126.rolling(window=126).std() # standard deviation over 126 trading days
        df[i+'_Year_momentum'] = returns252/stds252 # momentum ratio for 252 trading days
        df[i+'_6_month_momentum'] = returns126/stds126 # momentum ratio for 126 trading days
    df.dropna(inplace=True) # drop rows with NaN values
    return df

momentum_df = momentum_ratio(sd1)
```

```python
def Calculate_Z_Score(df,columns=sd1_columns):
    for i in columns:  # calculate z score for each stock
        mean252 = df[i+'_Year_momentum'].rolling(window=252).mean() # mean over 252 trading days
        std252 = df[i+'_Year_momentum'].rolling(window=252).std() # standard deviation over 252 trading days
        df[i+'Year_Z_Score'] = (df[i]-mean252)/std252 # z score for 252 trading days
        mean126 = df[i+'_6_month_momentum'].rolling(window=126).mean() # mean over 126 trading days
        std126 = df[i+'_6_month_momentum'].rolling(window=126).std() # standard deviation over 126 trading days
        df[i+'6_month_Z_Score'] = (df[i]-mean126)/std126 # z score for 126 trading days
        df[i+'ALL_Z_Score'] = (df[i+'Year_Z_Score'] + df[i+'6_month_Z_Score'])/2 # average z score (weighted average)
    df.dropna(inplace=True) # drop rows with NaN values
    return df
momentum_df = Calculate_Z_Score(momentum_df) # calculate z score
momentum_df
```
